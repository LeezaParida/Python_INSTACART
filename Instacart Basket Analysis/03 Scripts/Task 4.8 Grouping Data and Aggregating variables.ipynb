## Q4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# create new column to store max order to identify loyalty customers
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# setting loyalty flag as 'loyal customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# setting loyalty flag as 'regular customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# setting loyalty flag as 'new customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# frequency of your new “loyalty_flag” column
ords_prods_merge['loyalty_flag'].value_counts()
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(10)
```

## Q6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria: If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.” If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# create new column to store average of price to identify types of spenders
ords_prods_merge['avg_prices'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
# setting spending flag as 'low spender'
ords_prods_merge.loc[ords_prods_merge['avg_prices'] <10, 'spending_flag'] = 'Low Spender'
```

```python
# setting spending flag as 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_prices'] >=10, 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge['spending_flag'].value_counts()
```

```python
ords_prods_merge[['user_id', 'spending_flag', 'avg_prices']].head(10)
```

## Q7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows: If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.” If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.” If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# create new column to store median of 'days_since_prior_order' to identify frequent customer
ords_prods_merge['median_dspo'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(10)
```

```python
# setting order_frequency flag as 'Non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_dspo'] >20, 'order_frequency'] = 'Non-frequent customer'
```

```python
# setting order_frequency flag as 'Regular customer'
ords_prods_merge.loc[(ords_prods_merge['median_dspo'] >10) & (ords_prods_merge['median_dspo'] < 20), 'order_frequency'] = 'Regular customer'
```

```python
# setting order_frequency flag as 'Frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_dspo'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
ords_prods_merge['order_frequency'].value_counts()
```

```python
ords_prods_merge[['user_id', 'order_frequency', 'median_dspo']].head(10)
```
